```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable
from matplotlib.colors import Normalize

# -----------------------------
# progress bar (tqdm) with safe fallback
# -----------------------------
try:
    from tqdm.auto import tqdm
except Exception:
    def tqdm(x, **kwargs):
        return x

# ------------------------------------------------------------
# ASSUMES you already have:
#   - step_cosine_and_magnitude(X, ...)  (your updated version with normalize_by_n)
#   - trace_l, dff_l, coord_l, bind_l
#   - adfn.save_name
#   - tfn.select_region
# ------------------------------------------------------------


def null_time_shuffle_means(
    X: np.ndarray,
    n_shuffles: int = 200,
    seed: int = 0,
    mag: str = "l2",
    nan_on_zeros: bool = True,
    normalize_by_n: bool = True,
):
    rng = np.random.default_rng(seed)
    N, T = X.shape
    if T < 3:
        raise ValueError(f"Need T>=3; got T={T}")

    cos_obs, mag_obs = step_cosine_and_magnitude(
        X, mag=mag, nan_on_zeros=nan_on_zeros, normalize_by_n=normalize_by_n
    )
    obs = {"cos_mean": float(np.nanmean(cos_obs)), "mag_mean": float(np.nanmean(mag_obs))}

    cos_null = np.empty(n_shuffles, dtype=float)
    mag_null = np.empty(n_shuffles, dtype=float)

    for s in range(n_shuffles):
        perm = rng.permutation(T)
        Xs = X[:, perm]
        cos_s, mag_s = step_cosine_and_magnitude(
            Xs, mag=mag, nan_on_zeros=nan_on_zeros, normalize_by_n=normalize_by_n
        )
        cos_null[s] = np.nanmean(cos_s)
        mag_null[s] = np.nanmean(mag_s)

    return obs, {"cos_mean": cos_null, "mag_mean": mag_null}


def null_circular_per_neuron_means(
    X: np.ndarray,
    n_shuffles: int = 200,
    seed: int = 0,
    mag: str = "l2",
    nan_on_zeros: bool = True,
    normalize_by_n: bool = True,
):
    rng = np.random.default_rng(seed)
    N, T = X.shape
    if T < 3:
        raise ValueError(f"Need T>=3; got T={T}")

    cos_obs, mag_obs = step_cosine_and_magnitude(
        X, mag=mag, nan_on_zeros=nan_on_zeros, normalize_by_n=normalize_by_n
    )
    obs = {"cos_mean": float(np.nanmean(cos_obs)), "mag_mean": float(np.nanmean(mag_obs))}

    base = np.arange(T)
    cos_null = np.empty(n_shuffles, dtype=float)
    mag_null = np.empty(n_shuffles, dtype=float)

    for s in range(n_shuffles):
        shifts = rng.integers(0, T, size=N)
        idx = (base[None, :] - shifts[:, None]) % T     # (N,T)
        Xs = X[np.arange(N)[:, None], idx]              # (N,T)

        cos_s, mag_s = step_cosine_and_magnitude(
            Xs, mag=mag, nan_on_zeros=nan_on_zeros, normalize_by_n=normalize_by_n
        )
        cos_null[s] = np.nanmean(cos_s)
        mag_null[s] = np.nanmean(mag_s)

    return obs, {"cos_mean": cos_null, "mag_mean": mag_null}


def build_null_df(
    dff_l, trace_l, coord_l, bind_l,
    region="Telencephalon",
    n_shuffles=200,
    seed=0,
    mag="l2",
    show_progress: bool = True,
):
    rows = []

    it = range(len(dff_l))
    if show_progress:
        it = tqdm(it, desc=f"Nulls ({region})", total=len(dff_l), leave=True)

    for i in it:
        trace = np.load(trace_l[i])
        dff   = np.load(dff_l[i])
        coord = np.load(coord_l[i])
        bind  = np.load(bind_l[i])

        name = adfn.save_name(dff_l[i])
        try:
            age = os.path.basename(dff_l[i]).split("dpf")[0][-1]
        except Exception:
            age = np.nan
        age_num = pd.to_numeric(age, errors="coerce")

        _, sub_dff, _, _ = tfn.select_region(trace, dff, bind, coord, region)
        X = sub_dff

        # ---- time shuffle ----
        obs_ts, null_ts = null_time_shuffle_means(
            X, n_shuffles=n_shuffles, seed=seed + 1000*i, mag=mag,
            nan_on_zeros=True, normalize_by_n=True
        )

        # ---- circular per-neuron ----
        obs_c, null_c = null_circular_per_neuron_means(
            X, n_shuffles=n_shuffles, seed=seed + 1000*i + 7, mag=mag,
            nan_on_zeros=True, normalize_by_n=True
        )

        # observed rows
        for method, obs in [("time_shuffle", obs_ts), ("circular_per_neuron", obs_c)]:
            rows.append({
                "i": i, "name": name, "age": age, "age_num": age_num, "region": region,
                "method": method, "kind": "obs", "shuffle": -1,
                "cos_mean": obs["cos_mean"], "mag_mean": obs["mag_mean"],
                "n_neurons": int(X.shape[0]), "T": int(X.shape[1]),
            })

        # null rows
        for s in range(n_shuffles):
            rows.append({
                "i": i, "name": name, "age": age, "age_num": age_num, "region": region,
                "method": "time_shuffle", "kind": "null", "shuffle": s,
                "cos_mean": float(null_ts["cos_mean"][s]),
                "mag_mean": float(null_ts["mag_mean"][s]),
                "n_neurons": int(X.shape[0]), "T": int(X.shape[1]),
            })
            rows.append({
                "i": i, "name": name, "age": age, "age_num": age_num, "region": region,
                "method": "circular_per_neuron", "kind": "null", "shuffle": s,
                "cos_mean": float(null_c["cos_mean"][s]),
                "mag_mean": float(null_c["mag_mean"][s]),
                "n_neurons": int(X.shape[0]), "T": int(X.shape[1]),
            })

        # optional: update tqdm postfix with quick stats
        if show_progress and hasattr(it, "set_postfix"):
            it.set_postfix({
                "i": i,
                "age": age,
                "N": int(X.shape[0]),
                "cos_obs": f"{obs_ts['cos_mean']:.3g}",
            })

    df_long = pd.DataFrame(rows)
    df_long = df_long[np.isfinite(df_long["age_num"])].copy()
    df_long["age_num"] = df_long["age_num"].astype(float)
    return df_long


# -----------------------------
# Usage
# -----------------------------
df_null_long = build_null_df(dff_l, trace_l, coord_l, bind_l,
                            region="Telencephalon", n_shuffles=200, seed=0, mag="l2",
                            show_progress=True)

```

```python
import numpy as np
import matplotlib.pyplot as plt

# df_long = your dataframe shown above
# Make sure it's numeric
df_long = df_null_long 
df_long["age_num"] = pd.to_numeric(df_long["age_num"], errors="coerce")
df_long["cos_mean"] = pd.to_numeric(df_long["cos_mean"], errors="coerce")

def plot_age_vs_cos_by_method(df_long, include_null=True, max_null_points=30000, jitter=0.10):
    rng = np.random.default_rng(0)

    methods = list(df_long["method"].dropna().unique())
    methods = [m for m in methods if m in ("time_shuffle", "circular_per_neuron")] + \
              [m for m in methods if m not in ("time_shuffle", "circular_per_neuron")]

    for method in methods:
        dfm = df_long[(df_long["method"] == method) &
                      np.isfinite(df_long["age_num"]) &
                      np.isfinite(df_long["cos_mean"])].copy()

        df_obs  = dfm[dfm["kind"] == "obs"].copy()
        df_null = dfm[dfm["kind"] == "null"].copy()

        # downsample nulls for plotting density
        if include_null and len(df_null) > max_null_points:
            df_null = df_null.sample(n=max_null_points, random_state=0)

        plt.figure(figsize=(8, 5))

        # null: faint cloud
        if include_null and len(df_null) > 0:
            x_null = df_null["age_num"].to_numpy() + rng.uniform(-jitter, jitter, size=len(df_null))
            plt.scatter(x_null, df_null["cos_mean"].to_numpy(), s=10, alpha=0.06, linewidths=0)

        # obs: bold points
        x_obs = df_obs["age_num"].to_numpy() + rng.uniform(-jitter, jitter, size=len(df_obs))
        plt.scatter(x_obs, df_obs["cos_mean"].to_numpy(), s=60, alpha=0.95, edgecolor="k", linewidths=0.3)

        # per-age means (null + obs)
        ages = np.sort(df_obs["age_num"].unique())
        if include_null and len(df_null) > 0:
            null_mean = df_null.groupby("age_num")["cos_mean"].mean().reindex(ages)
            plt.plot(ages, null_mean.to_numpy(), linestyle="--", linewidth=2)

        obs_mean = df_obs.groupby("age_num")["cos_mean"].mean().reindex(ages)
        plt.plot(ages, obs_mean.to_numpy(), linestyle="-", linewidth=2)

        plt.xlabel("Age (dpf)")
        plt.ylabel("cos_mean (t→t+1)")
        plt.yscale('log')
        plt.title(f"Age vs cos_mean — {method}\nobs=bold, null=faint, mean(obs)=solid, mean(null)=dashed")
        plt.tight_layout()
        plt.show()


# run
plot_age_vs_cos_by_method(df_long, include_null=True, max_null_points=25000, jitter=0.10)
```

```python
import os
import numpy as np
import pandas as pd

try:
    from tqdm.auto import tqdm
except Exception:
    def tqdm(x, **kwargs):
        return x


# -----------------------------
# Null: global time shuffle (same perm for all neurons)
# -----------------------------
def null_global_time_shuffle_means(
    X: np.ndarray,
    n_shuffles: int = 200,
    seed: int = 0,
    mag: str = "l2",
    nan_on_zeros: bool = True,
    normalize_by_n: bool = True,
):
    """
    Global time shuffle null:
      - applies ONE permutation of time indices to all neurons simultaneously
      - preserves each population snapshot (cross-neuron correlations at a timepoint)
      - destroys temporal adjacency/continuity
    """
    rng = np.random.default_rng(seed)
    N, T = X.shape
    if T < 3:
        raise ValueError(f"Need T>=3; got T={T}")

    cos_obs, mag_obs = step_cosine_and_magnitude(
        X, mag=mag, nan_on_zeros=nan_on_zeros, normalize_by_n=normalize_by_n
    )
    obs = {"cos_mean": float(np.nanmean(cos_obs)), "mag_mean": float(np.nanmean(mag_obs))}

    cos_null = np.empty(n_shuffles, dtype=float)
    mag_null = np.empty(n_shuffles, dtype=float)

    for s in range(n_shuffles):
        perm = rng.permutation(T)
        Xs = X[:, perm]  # same perm for all neurons
        cos_s, mag_s = step_cosine_and_magnitude(
            Xs, mag=mag, nan_on_zeros=nan_on_zeros, normalize_by_n=normalize_by_n
        )
        cos_null[s] = float(np.nanmean(cos_s))
        mag_null[s] = float(np.nanmean(mag_s))

    return obs, {"cos_mean": cos_null, "mag_mean": mag_null}


# -----------------------------
# Build DF across sessions (same schema as your df_null_long)
# -----------------------------
def build_global_time_shuffle_df(
    dff_l, trace_l, coord_l, bind_l,
    region: str = "Telencephalon",
    n_shuffles: int = 200,
    seed: int = 0,
    mag: str = "l2",
    show_progress: bool = True,
):
    rows = []

    it = range(len(dff_l))
    if show_progress:
        it = tqdm(it, desc=f"Global time-shuffle nulls ({region})", total=len(dff_l), leave=True)

    for i in it:
        # ---- load ----
        trace = np.load(trace_l[i])
        dff   = np.load(dff_l[i])
        coord = np.load(coord_l[i])
        bind  = np.load(bind_l[i])

        name = adfn.save_name(dff_l[i])
        try:
            age = os.path.basename(dff_l[i]).split("dpf")[0][-1]
        except Exception:
            age = np.nan
        age_num = pd.to_numeric(age, errors="coerce")

        # ---- region select ----
        _, sub_dff, _, _ = tfn.select_region(trace, dff, bind, coord, region)
        X = sub_dff

        # ---- compute obs + null ----
        obs_g, null_g = null_global_time_shuffle_means(
            X,
            n_shuffles=n_shuffles,
            seed=seed + 1000*i,          # deterministic per session
            mag=mag,
            nan_on_zeros=True,
            normalize_by_n=True,
        )

        # observed row
        rows.append({
            "i": i, "name": name, "age": age, "age_num": age_num, "region": region,
            "method": "global_time_shuffle", "kind": "obs", "shuffle": -1,
            "cos_mean": obs_g["cos_mean"], "mag_mean": obs_g["mag_mean"],
            "n_neurons": int(X.shape[0]), "T": int(X.shape[1]),
        })

        # null rows
        for s in range(n_shuffles):
            rows.append({
                "i": i, "name": name, "age": age, "age_num": age_num, "region": region,
                "method": "global_time_shuffle", "kind": "null", "shuffle": s,
                "cos_mean": float(null_g["cos_mean"][s]),
                "mag_mean": float(null_g["mag_mean"][s]),
                "n_neurons": int(X.shape[0]), "T": int(X.shape[1]),
            })

        # optional tqdm postfix
        if show_progress and hasattr(it, "set_postfix"):
            it.set_postfix({
                "i": i,
                "age": age,
                "N": int(X.shape[0]),
                "cos_obs": f"{obs_g['cos_mean']:.3g}",
            })

    df_long = pd.DataFrame(rows)
    df_long = df_long[np.isfinite(df_long["age_num"])].copy()
    df_long["age_num"] = df_long["age_num"].astype(float)
    return df_long


# -----------------------------
# Usage
# -----------------------------
df_global_shuffle = build_global_time_shuffle_df(
    dff_l, trace_l, coord_l, bind_l,
    region="Telencephalon",
    n_shuffles=200,
    seed=0,
    mag="l2",
    show_progress=True
)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# expects: df_global_shuffle (from build_global_time_shuffle_df)
# columns: age_num, kind in {"obs","null"}, cos_mean, mag_mean, shuffle, i, name, method

def plot_obs_vs_null_by_age(
    df_long: pd.DataFrame,
    metric: str = "cos_mean",
    method: str = "global_time_shuffle",
    max_null_points: int = 30000,
    jitter: float = 0.10,
    seed: int = 0,
):
    rng = np.random.default_rng(seed)

    df = df_long.copy()
    df["age_num"] = pd.to_numeric(df["age_num"], errors="coerce")
    df[metric] = pd.to_numeric(df[metric], errors="coerce")
    df = df[np.isfinite(df["age_num"]) & np.isfinite(df[metric])].copy()

    if "method" in df.columns:
        df = df[df["method"] == method].copy()

    df_obs  = df[df["kind"] == "obs"].copy()
    df_null = df[df["kind"] == "null"].copy()

    # downsample nulls for plotting (keeps UI snappy)
    if len(df_null) > max_null_points:
        df_null = df_null.sample(n=max_null_points, random_state=seed)

    # jitter x for visibility
    x_obs  = df_obs["age_num"].to_numpy()  + rng.uniform(-jitter, jitter, size=len(df_obs))
    x_null = df_null["age_num"].to_numpy() + rng.uniform(-jitter, jitter, size=len(df_null))

    plt.figure(figsize=(8, 5))

    # null cloud (faint)
    if len(df_null) > 0:
        plt.scatter(x_null, df_null[metric].to_numpy(), s=10, alpha=0.06, linewidths=0, label="null (samples)")

    # obs points (bold)
    plt.scatter(x_obs, df_obs[metric].to_numpy(), s=70, alpha=0.9, edgecolor="k", linewidths=0.3, label="obs (session mean)")

    # per-age means (null + obs)
    ages = np.sort(df_obs["age_num"].dropna().unique())
    obs_mean  = df_obs.groupby("age_num")[metric].mean().reindex(ages)
    plt.plot(ages, obs_mean.to_numpy(), linewidth=2, label="mean(obs)")

    if len(df_null) > 0:
        null_mean = df_null.groupby("age_num")[metric].mean().reindex(ages)
        plt.plot(ages, null_mean.to_numpy(), linestyle="--", linewidth=2, label="mean(null)")

    plt.xlabel("Age (dpf)")
    plt.ylabel(metric)
    plt.title(f"{method}: age vs {metric}\nobs=bold, null=faint; mean(obs)=solid, mean(null)=dashed")
    plt.legend(frameon=False)
    plt.tight_layout()
    plt.show()


def plot_null_corrected_delta(
    df_long: pd.DataFrame,
    metric: str = "cos_mean",
    method: str = "global_time_shuffle",
):
    """
    Plots delta(age) = obs - mean(null) per session and its per-age mean.
    """
    df = df_long.copy()
    df["age_num"] = pd.to_numeric(df["age_num"], errors="coerce")
    df[metric] = pd.to_numeric(df[metric], errors="coerce")
    df = df[np.isfinite(df["age_num"]) & np.isfinite(df[metric])].copy()

    if "method" in df.columns:
        df = df[df["method"] == method].copy()

    obs = (df[df["kind"] == "obs"]
           .groupby(["i", "age_num"], as_index=False)[metric].mean()
           .rename(columns={metric: "obs"}))

    nul = (df[df["kind"] == "null"]
           .groupby(["i", "age_num"], as_index=False)[metric].mean()
           .rename(columns={metric: "null_mean"}))

    m = obs.merge(nul, on=["i", "age_num"], how="inner")
    m["delta"] = m["obs"] - m["null_mean"]

    ages = np.sort(m["age_num"].unique())
    by_age = m.groupby("age_num", as_index=False)["delta"].mean().sort_values("age_num")

    plt.figure(figsize=(8, 5))
    plt.scatter(m["age_num"].to_numpy(), m["delta"].to_numpy(), s=70, alpha=0.85, edgecolor="k", linewidths=0.3)
    plt.plot(by_age["age_num"].to_numpy(), by_age["delta"].to_numpy(), linewidth=2)
    plt.axhline(0, linestyle="--", linewidth=1)
    plt.xlabel("Age (dpf)")
    plt.ylabel(f"delta = obs - mean(null) [{metric}]")
    plt.title(f"{method}: null-corrected delta vs age")
    plt.tight_layout()
    plt.show()

    return m


# ---- run the plots ----
plot_obs_vs_null_by_age(df_global_shuffle, metric="cos_mean", method="global_time_shuffle")
plot_null_corrected_delta(df_global_shuffle, metric="cos_mean", method="global_time_shuffle")

# (optional) magnitude too
# plot_obs_vs_null_by_age(df_global_shuffle, metric="mag_mean", method="global_time_shuffle")
# plot_null_corrected_delta(df_global_shuffle, metric="mag_mean", method="global_time_shuffle")
```

```python
import os
import numpy as np
import pandas as pd

try:
    from tqdm.auto import tqdm
except Exception:
    def tqdm(x, **kwargs):
        return x


def lag1_autocorr_per_neuron_fast(X: np.ndarray, eps: float = 1e-12) -> np.ndarray:
    """
    Fast lag-1 autocorr per neuron.
    X: (N, T)
    Returns r: (N,) where r[n] = corr(X[n, :-1], X[n, 1:])
    - Handles constant/near-constant traces -> NaN.
    """
    X0 = X[:, :-1]
    X1 = X[:, 1:]

    # mean-center per neuron
    m0 = X0.mean(axis=1, keepdims=True)
    m1 = X1.mean(axis=1, keepdims=True)
    Y0 = X0 - m0
    Y1 = X1 - m1

    # covariance and stds (vectorised)
    cov = np.einsum("ij,ij->i", Y0, Y1) / (X0.shape[1] - 1)
    v0  = np.einsum("ij,ij->i", Y0, Y0) / (X0.shape[1] - 1)
    v1  = np.einsum("ij,ij->i", Y1, Y1) / (X0.shape[1] - 1)

    denom = np.sqrt(v0 * v1)
    r = np.full(X.shape[0], np.nan, dtype=np.float32)
    ok = denom > eps
    r[ok] = (cov[ok] / denom[ok]).astype(np.float32)

    # numerical safety
    r = np.clip(r, -1.0, 1.0)
    return r


rows = []

for i in tqdm(range(len(dff_l)), total=len(dff_l), desc="lag1 autocorr (Telencephalon)"):
    trace = np.load(trace_l[i])
    dff   = np.load(dff_l[i])
    coord = np.load(coord_l[i])
    bind  = np.load(bind_l[i])

    name = adfn.save_name(dff_l[i])
    try:
        age = os.path.basename(dff_l[i]).split("dpf")[0][-1]
    except Exception:
        age = np.nan
    age_num = pd.to_numeric(age, errors="coerce")

    _, sub_dff, _, _ = tfn.select_region(trace, dff, bind, coord, "Telencephalon")
    X = sub_dff  # (N,T)

    r = lag1_autocorr_per_neuron_fast(X)

    rows.append({
        "i": i,
        "name": name,
        "age": age,
        "age_num": age_num,
        "region": "Telencephalon",
        "n_neurons": int(X.shape[0]),
        "T": int(X.shape[1]),

        # session summaries
        "lag1_mean": float(np.nanmean(r)),
        "lag1_median": float(np.nanmedian(r)),
        "lag1_std": float(np.nanstd(r)),
        "lag1_q05": float(np.nanquantile(r, 0.05)),
        "lag1_q95": float(np.nanquantile(r, 0.95)),
        "lag1_nan_frac": float(np.mean(~np.isfinite(r))),
    })

df_lag1 = pd.DataFrame(rows).sort_values(["age_num", "name", "i"]).reset_index(drop=True)
df_lag1
```

```python
import numpy as np
import matplotlib.pyplot as plt

def step_cosine_and_magnitude(
    X: np.ndarray,
    eps: float = 1e-12,
    mag: str = "l2",             # "l2" or "l1"
    nan_on_zeros: bool = True,   # if a(t) or a(t+1) is all-zeros
    normalize_by_n: bool = True  # normalize magnitude by neuron count
):
    """
    X: array with shape (n_neurons, n_time)

    Returns:
      cos_dist: (T-1,) where cos_dist[t] = 1 - cos_sim(X[:,t], X[:,t+1])
      mag_dist: (T-1,) magnitude of delta per step, optionally normalized:
          - L2: ||Δ||2 / sqrt(N)  (RMS per neuron)
          - L1: ||Δ||1 / N        (mean abs per neuron)
    """
    assert X.ndim == 2, f"Expected 2D (neurons,time), got {X.shape}"
    N, T = X.shape
    if T < 2:
        raise ValueError(f"Need at least 2 timepoints, got T={T}")

    # time-major views: (T-1, N)
    A = X[:, :-1].T
    B = X[:, 1: ].T

    # ---- cosine distance ----
    dot = np.einsum("ij,ij->i", A, B)        # (T-1,)
    nA  = np.linalg.norm(A, axis=1)          # (T-1,)
    nB  = np.linalg.norm(B, axis=1)          # (T-1,)
    denom = nA * nB

    if nan_on_zeros:
        cos_sim = np.full_like(dot, np.nan, dtype=float)
        ok = denom > eps
        cos_sim[ok] = dot[ok] / denom[ok]
    else:
        cos_sim = dot / (denom + eps)

    cos_sim = np.clip(cos_sim, -1.0, 1.0)
    cos_dist = 1.0 - cos_sim

    # ---- magnitude distance (normalized per neuron if requested) ----
    D = B - A  # (T-1, N)

    mag = mag.lower()
    if mag == "l2":
        mag_dist = np.linalg.norm(D, axis=1)  # ||Δ||2
        if normalize_by_n:
            mag_dist = mag_dist / np.sqrt(N)  # RMS per neuron
    elif mag == "l1":
        mag_dist = np.sum(np.abs(D), axis=1)  # ||Δ||1
        if normalize_by_n:
            mag_dist = mag_dist / N           # mean abs per neuron
    else:
        raise ValueError("mag must be 'l2' or 'l1'")

    return cos_dist, mag_dist


# =========================
# Run on your telencephalon data
# =========================
X = sub_dff  # (neurons, time)

cos_d, mag_d = step_cosine_and_magnitude(X, mag="l2", nan_on_zeros=True, normalize_by_n=True)

print("cos_d:", cos_d.shape, "mag_d:", mag_d.shape)
print("cos_d mean±sd:", np.nanmean(cos_d), np.nanstd(cos_d))
print("mag_d mean±sd:", np.nanmean(mag_d), np.nanstd(mag_d))
print(name)

plt.figure()
plt.plot(cos_d)
plt.title("t → t+1 Cosine distance (Telencephalon, dFF)")
plt.xlabel("t")
plt.ylabel("1 - cosine_similarity")
plt.show()

plt.figure()
plt.plot(mag_d)
plt.title("t → t+1 Magnitude distance (RMS per neuron) (Telencephalon, dFF)")
plt.xlabel("t")
plt.ylabel("||x(t+1) - x(t)||2 / sqrt(N)")
plt.show()
```

```python
import os, re, glob
import numpy as np
import pandas as pd
import powerlaw
from tqdm import tqdm

# -----------------------
# parse helpers
# -----------------------
def parse_age_dpf(path):
    m = re.search(r"(\d+)\s*dpf", path)
    return int(m.group(1)) if m else np.nan

def parse_animal_id(path):
    base = os.path.basename(path)
    m = re.search(r"(BLNDEV-WILDTYPE-\d+)", base)
    return m.group(1) if m else base.split("_")[0]

# -----------------------
# powerlaw exponent with FIXED xmin, NO xmax
# -----------------------
def pl_alpha_fixedxmin(x, xmin, discrete=True):
    x = np.asarray(x).ravel()
    x = x[np.isfinite(x) & (x > 0)]
    # IMPORTANT: keep integers for discrete fit
    if discrete:
        x = x.astype(int)
    x = x[x >= xmin]
    if x.size < 50:
        return np.nan
    fit = powerlaw.Fit(x, discrete=discrete, xmin=xmin, verbose=False)
    return float(fit.power_law.alpha)

# -----------------------
# DCC exactly like yours, but exponents are provided (from powerlaw)
# -----------------------
def dcc_like_yours_from_exps(av_size, av_dur, size_exp, dur_exp, min_size=3, dur_scale=2.73):
    S = np.asarray(av_size).ravel().astype(float)
    T = np.asarray(av_dur).ravel().astype(float)
    m = np.isfinite(S) & np.isfinite(T) & (S > 0) & (T > 0)
    S, T = S[m], T[m]

    # match your avalanche() cut
    keep = S >= min_size
    S, T = S[keep], T[keep]
    if S.size < 50:
        return np.nan, np.nan

    # match your duration scaling
    T = T / dur_scale

    # mean size per unique duration
    Tu = np.unique(T)
    size_vec = np.array([np.mean(S[T == t]) for t in Tu], float)
    xaxis = Tu.astype(float)
    yaxis = size_vec.astype(float)

    # match your "drop last point" behavior
    if xaxis.size < 5:
        return np.nan, np.nan
    xfit = xaxis[:-1]
    yfit = yaxis[:-1]

    ok = (xfit > 0) & (yfit > 0) & np.isfinite(xfit) & np.isfinite(yfit)
    xfit, yfit = xfit[ok], yfit[ok]
    if xfit.size < 5:
        return np.nan, np.nan

    fit_beta, _ = np.polyfit(np.log10(xfit), np.log10(yfit), 1)

    if not (np.isfinite(size_exp) and np.isfinite(dur_exp)) or (size_exp == 1.0):
        return np.nan, float(fit_beta)

    pred_beta = (dur_exp - 1.0) / (size_exp - 1.0)
    dcc = float(np.abs(fit_beta - pred_beta))
    return dcc, float(fit_beta)

# -----------------------
# main loop: powerlaw exps + your DCC recipe (NO xmax)
# -----------------------
fdata = "/Users/k2585057/Library/CloudStorage/Dropbox/PhD/Analysis/Project/BLNDEV-WILDTYPE"
av_l  = np.sort(glob.glob(f"{fdata}/*Tel*crit*"))

rows = []
for _, fp in tqdm(enumerate(av_l), total=len(av_l)):
    d = np.load(fp, allow_pickle=True).item()
    sizes = np.asarray(d["av"][0]).ravel()
    durs  = np.asarray(d["av"][1]).ravel()

    # match your xmin choices
    xmin_size = int(np.nanmin(sizes[np.isfinite(sizes) & (sizes > 0)]))  # often 3
    xmin_dur  = 2

    # exponents via powerlaw (discrete, fixed xmin, no xmax)
    tau_size  = pl_alpha_fixedxmin(sizes, xmin=xmin_size, discrete=True)
    alpha_dur = pl_alpha_fixedxmin(durs,  xmin=xmin_dur,  discrete=True)

    # DCC via your recipe, using these exponents
    dcc, fit_beta = dcc_like_yours_from_exps(
        sizes, durs, tau_size, alpha_dur, min_size=3, dur_scale=2.73
    )

    rows.append({
        "file": fp,
        "animal": parse_animal_id(fp),
        "age_dpf": parse_age_dpf(fp),
        "region": "Telencephalon" if "Telencephalon" in fp else np.nan,
        "n_avalanches": int(min(len(sizes), len(durs))),

        "xmin_size": xmin_size,
        "xmin_dur": xmin_dur,
        "tau_size": tau_size,
        "alpha_dur": alpha_dur,
        "fit_beta": fit_beta,
        "dcc": dcc,
    })

df_out = pd.DataFrame(rows).sort_values(["age_dpf", "animal"]).reset_index(drop=True)

out_csv = os.path.join(fdata, "dcc_powerlaw_fixedxmin_NOxmax.csv")
df_out.to_csv(out_csv, index=False)
print(df_out.head())
print("Saved:", out_csv)
```

```python
#estimate timescale of each neuron
def tau_half_life_from_acf_fast(
    X: np.ndarray,
    dt: float,
    max_lag_s: float = 10.0,
    threshold: float = 0.5,      # use 0.5 for half-life, or 1/np.e for e-fold
    eps: float = 1e-12,
) -> tuple[np.ndarray, np.ndarray]:
    """
    Option B: nonparametric timescale per neuron from the autocorrelation function (ACF).

    Returns:
      tau: (N,) timescale in seconds defined as the first lag where ACF <= threshold.
           If never crosses within max_lag, returns max_lag_s (censored).
           If lag-0 is undefined (constant trace), returns NaN.
      acf1: (N,) ACF at lag 1 (useful sanity check)

    Fast vectorised ACF over lags 0..L using dot products.

    X: (N,T)
    dt: seconds per frame
    max_lag_s: compute up to this lag in seconds
    threshold: crossing threshold (0.5 half-life, or 1/np.e e-fold)
    """
    assert X.ndim == 2, f"Expected (N,T), got {X.shape}"
    N, T = X.shape
    L = int(np.floor(max_lag_s / dt))
    L = max(1, min(L, T - 2))  # at least 1, and feasible

    # mean-center per neuron
    Xc = X - X.mean(axis=1, keepdims=True)

    # variance per neuron (lag 0 autocov)
    var = np.einsum("ij,ij->i", Xc, Xc) / max(T - 1, 1)
    ok = var > eps

    # compute ACF for lags 0..L
    # acf[k] = cov(x(t), x(t+k)) / var
    acf = np.full((N, L + 1), np.nan, dtype=np.float32)
    if np.any(ok):
        Xok = Xc[ok]
        var_ok = var[ok]

        # lag 0
        acf_ok = np.empty((Xok.shape[0], L + 1), dtype=np.float32)
        acf_ok[:, 0] = 1.0

        # lags 1..L
        # dot across time with shifted versions
        for k in range(1, L + 1):
            a = Xok[:, :-k]
            b = Xok[:,  k:]
            cov = np.einsum("ij,ij->i", a, b) / max((T - k - 1), 1)
            acf_ok[:, k] = (cov / (var_ok + eps)).astype(np.float32)

        acf[ok] = acf_ok

    # lag-1 acf
    acf1 = acf[:, 1].copy()

    # find first crossing <= threshold (ignoring lag 0)
    tau = np.full(N, np.nan, dtype=np.float32)
    if np.any(ok):
        crosses = (acf[ok, 1:] <= threshold) & np.isfinite(acf[ok, 1:])
        # argmax gives first True only if any True; handle "never crosses"
        any_cross = crosses.any(axis=1)
        first_idx = np.argmax(crosses, axis=1)  # index in 0..L-1
        lag_frames = np.where(any_cross, first_idx + 1, L).astype(np.float32)  # +1 because we dropped lag0
        tau[ok] = lag_frames * float(dt)

    return tau, acf1


# ---- usage ----
# dt = 1.0 / frame_rate_hz
# tau_half, acf1 = tau_half_life_from_acf_fast(sub_dff, dt=dt, max_lag_s=10.0, threshold=0.5)
# tau_efold, _   = tau_half_life_from_acf_fast(sub_dff, dt=dt, max_lag_s=10.0, threshold=1/np.e)
```

```python
import os
import numpy as np
import pandas as pd

try:
    from tqdm.auto import tqdm
except Exception:
    def tqdm(x, **kwargs):
        return x


# -----------------------------
# Option B: timescale via ACF threshold crossing (per neuron)
# -----------------------------
def tau_half_life_from_acf_fast(
    X: np.ndarray,
    dt: float,
    max_lag_s: float = 10.0,
    threshold: float = 0.5,      # 0.5 = half-life, 1/np.e = e-fold
    eps: float = 1e-12,
) -> tuple[np.ndarray, np.ndarray]:
    assert X.ndim == 2, f"Expected (N,T), got {X.shape}"
    N, T = X.shape
    L = int(np.floor(max_lag_s / dt))
    L = max(1, min(L, T - 2))

    Xc = X - X.mean(axis=1, keepdims=True)

    # variance per neuron
    var = np.einsum("ij,ij->i", Xc, Xc) / max(T - 1, 1)
    ok = var > eps

    acf = np.full((N, L + 1), np.nan, dtype=np.float32)
    acf[:, 0] = 1.0

    if np.any(ok):
        Xok = Xc[ok]
        var_ok = var[ok]

        acf_ok = np.empty((Xok.shape[0], L + 1), dtype=np.float32)
        acf_ok[:, 0] = 1.0

        # lags 1..L
        for k in range(1, L + 1):
            a = Xok[:, :-k]
            b = Xok[:,  k:]
            cov = np.einsum("ij,ij->i", a, b) / max((T - k - 1), 1)
            acf_ok[:, k] = (cov / (var_ok + eps)).astype(np.float32)

        acf[ok] = acf_ok

    acf1 = acf[:, 1].copy()

    # first crossing (lag>=1) where acf <= threshold
    tau = np.full(N, np.nan, dtype=np.float32)
    if np.any(ok):
        crosses = (acf[ok, 1:] <= threshold) & np.isfinite(acf[ok, 1:])
        any_cross = crosses.any(axis=1)
        first_idx = np.argmax(crosses, axis=1)  # 0..L-1
        lag_frames = np.where(any_cross, first_idx + 1, L).astype(np.float32)
        tau[ok] = lag_frames * float(dt)

    return tau, acf1


# -----------------------------
# Run across all fish/sessions -> long dataframe (one row per neuron)
# -----------------------------
def build_neuron_timescale_df(
    dff_l, trace_l, coord_l, bind_l,
    region: str = "Telencephalon",
    dt: float = 0.2,             # <-- set correctly (seconds per frame)
    max_lag_s: float = 10.0,
    threshold: float = 0.5,
):
    rows = []

    for i in tqdm(range(len(dff_l)), total=len(dff_l), desc=f"Timescales ({region})"):
        trace = np.load(trace_l[i])
        dff   = np.load(dff_l[i])
        coord = np.load(coord_l[i])
        bind  = np.load(bind_l[i])

        name = adfn.save_name(dff_l[i])
        try:
            age = os.path.basename(dff_l[i]).split("dpf")[0][-1]
        except Exception:
            age = np.nan
        age_num = pd.to_numeric(age, errors="coerce")

        sub_trace, sub_dff, _, _ = tfn.select_region(trace, dff, bind, coord, region)
        X = sub_dff  # (N,T)

        tau, acf1 = tau_half_life_from_acf_fast(
            X, dt=dt, max_lag_s=max_lag_s, threshold=threshold
        )

        N, T = X.shape
        # append rows
        # neuron_idx is the index within the selected region (0..N-1)
        for n in range(N):
            rows.append({
                "i": i,
                "name": name,
                "age": age,
                "age_num": age_num,
                "region": region,
                "neuron_idx": n,
                "tau_s": float(tau[n]) if np.isfinite(tau[n]) else np.nan,
                "acf1":  float(acf1[n]) if np.isfinite(acf1[n]) else np.nan,
                "n_neurons": int(N),
                "T": int(T),
                "dt": float(dt),
                "max_lag_s": float(max_lag_s),
                "threshold": float(threshold),
            })

    df_neuron_tau = pd.DataFrame(rows)
    df_neuron_tau = df_neuron_tau[np.isfinite(df_neuron_tau["age_num"])].copy()
    df_neuron_tau["age_num"] = df_neuron_tau["age_num"].astype(float)
    return df_neuron_tau


# -----------------------------
# Usage (set dt correctly!)
# -----------------------------
# Example: if imaging at 5 Hz -> dt = 0.2
# dt = 1.0 / frame_rate_hz
df_neuron_tau = build_neuron_timescale_df(
    dff_l, trace_l, coord_l, bind_l,
    region="Telencephalon",
    dt=1/2.73,                 # <-- you must define dt
    max_lag_s=5,
    threshold=0.7
)

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

def plot_mean_tau_by_fish_age_with_spearman(
    df_neuron_tau: pd.DataFrame,
    region: str = "Telencephalon",
    use: str = "mean",          # "mean" or "median" (fish-level summary across neurons)
    show_censor: bool = True,
    censor_tol: float = 1e-6,
):
    df = df_neuron_tau.copy()
    df["age_num"] = pd.to_numeric(df["age_num"], errors="coerce")
    df["tau_s"]   = pd.to_numeric(df["tau_s"], errors="coerce")
    df = df[np.isfinite(df["age_num"]) & np.isfinite(df["tau_s"])].copy()

    if "region" in df.columns and region is not None:
        df = df[df["region"] == region].copy()

    # per-fish summary across neurons
    fish = (df.groupby(["name", "age_num"], as_index=False)
              .agg(tau_mean=("tau_s", "mean"),
                   tau_median=("tau_s", "median"),
                   n_neurons=("tau_s", "size")))

    # ---- censoring fraction (no apply) ----
    if show_censor and "max_lag_s" in df.columns:
        max_lag_s = pd.to_numeric(df["max_lag_s"], errors="coerce").dropna()
        max_lag_s = float(max_lag_s.iloc[0]) if len(max_lag_s) else None
    else:
        max_lag_s = None

    if show_censor and (max_lag_s is not None):
        is_cens = np.abs(df["tau_s"].to_numpy() - max_lag_s) <= censor_tol
        df["_is_cens"] = is_cens.astype(float)
        cens = (df.groupby(["name", "age_num"], as_index=False)
                  .agg(frac_censored=("_is_cens", "mean")))
        fish = fish.merge(cens, on=["name", "age_num"], how="left")
    else:
        fish["frac_censored"] = np.nan

    ycol = "tau_mean" if use.lower() == "mean" else "tau_median"

    # Spearman across fish-level points
    x = fish["age_num"].to_numpy(dtype=float)
    y = fish[ycol].to_numpy(dtype=float)
    m = np.isfinite(x) & np.isfinite(y)
    rho, p = spearmanr(x[m], y[m]) if m.sum() >= 3 else (np.nan, np.nan)

    # per-age mean of fish summaries
    age_summary = (fish.groupby("age_num", as_index=False)
                        .agg(mean_y=(ycol, "mean"),
                             sem_y=(ycol, lambda z: z.std(ddof=1)/np.sqrt(len(z)) if len(z)>1 else np.nan),
                             n_fish=(ycol, "size"),
                             mean_frac_censored=("frac_censored", "mean"))
                        .sort_values("age_num"))

    # plot
    plt.figure(figsize=(8, 5))
    plt.scatter(fish["age_num"].to_numpy(),
                fish[ycol].to_numpy(),
                s=70, alpha=0.85, edgecolor="k", linewidths=0.3, label=f"fish {use}")

    plt.errorbar(age_summary["age_num"].to_numpy(),
                 age_summary["mean_y"].to_numpy(),
                 yerr=age_summary["sem_y"].to_numpy(),
                 fmt='-o', capsize=3, label=f"age mean ± SEM (fish {use})")

    title = f"{region}: τ by fish and age (fish {use})\nSpearman ρ={rho:.3g}, p={p:.3g} (fish-level)"
    if show_censor and (max_lag_s is not None):
        title += f"\n(censor cap={max_lag_s:g}s; text labels=% censored per age)"
    plt.title(title)

    if show_censor and (max_lag_s is not None):
        for _, r in age_summary.iterrows():
            if np.isfinite(r["mean_frac_censored"]):
                plt.text(r["age_num"], r["mean_y"], f"{100*r['mean_frac_censored']:.0f}%",
                         fontsize=8, ha="left", va="bottom")

    plt.xlabel("Age (dpf)")
    plt.ylabel("Timescale τ (s) (ACF threshold-crossing)")
    plt.legend(frameon=False)
    plt.tight_layout()
    plt.show()

    return fish, age_summary, (rho, p)


# ---- run ----
fish_tau, age_tau, (rho, p) = plot_mean_tau_by_fish_age_with_spearman(
    df_neuron_tau, region="Telencephalon", use="mean", show_censor=True
)
print("Spearman (fish-level): rho =", rho, "p =", p)
```

```python
import os
import numpy as np
import pandas as pd

try:
    from tqdm.auto import tqdm
except Exception:
    def tqdm(x, **kwargs):
        return x


# -----------------------------
# Cosine + magnitude distance at arbitrary lag (t -> t+k)
# -----------------------------
def lag_cosine_and_magnitude(
    X: np.ndarray,
    k: int,
    eps: float = 1e-12,
    mag: str = "l2",                  # "l2" or "l1"
    nan_on_zeros: bool = True,
    normalize_by_n: bool = True,      # magnitude normalisation by neuron count
):
    """
    X: (N,T)
    k: lag in frames (>=1)
    Returns:
      cos_dist: (T-k,)  with cos_dist[t] = 1 - cos_sim(X[:,t], X[:,t+k])
      mag_dist: (T-k,)  with ||X[:,t+k]-X[:,t]|| (optionally normalised by N)
    """
    assert X.ndim == 2, f"Expected (N,T), got {X.shape}"
    N, T = X.shape
    if k < 1 or k >= T:
        raise ValueError(f"Invalid k={k} for T={T}")

    A = X[:, :-k].T   # (T-k, N)
    B = X[:,  k: ].T  # (T-k, N)

    # cosine distance
    dot   = np.einsum("ij,ij->i", A, B)
    nA    = np.linalg.norm(A, axis=1)
    nB    = np.linalg.norm(B, axis=1)
    denom = nA * nB

    if nan_on_zeros:
        cos_sim = np.full_like(dot, np.nan, dtype=float)
        ok = denom > eps
        cos_sim[ok] = dot[ok] / denom[ok]
    else:
        cos_sim = dot / (denom + eps)

    cos_sim  = np.clip(cos_sim, -1.0, 1.0)
    cos_dist = 1.0 - cos_sim

    # magnitude distance
    D = B - A
    if mag.lower() == "l2":
        mag_dist = np.linalg.norm(D, axis=1)
        if normalize_by_n:
            mag_dist = mag_dist / np.sqrt(N)
    elif mag.lower() == "l1":
        mag_dist = np.sum(np.abs(D), axis=1)
        if normalize_by_n:
            mag_dist = mag_dist / N
    else:
        raise ValueError("mag must be 'l2' or 'l1'")

    return cos_dist, mag_dist


# -----------------------------
# Run across all sessions, lags 1..10 seconds (1 s bins)
# -----------------------------
def build_future_lag_df(
    dff_l, trace_l, coord_l, bind_l,
    region: str = "Telencephalon",
    dt: float = None,          # seconds per frame (REQUIRED)
    max_lag_s: int = 10,       # up to 10 seconds
    step_s: int = 1,           # 1 second bins
    mag: str = "l2",
    nan_on_zeros: bool = True,
    normalize_by_n: bool = True,
):
    if dt is None:
        raise ValueError("Please pass dt (seconds per frame), e.g. dt = 1.0/frame_rate_hz")

    lag_secs = np.arange(step_s, max_lag_s + 1e-9, step_s, dtype=float)

    rows = []
    for i in tqdm(range(len(dff_l)), total=len(dff_l), desc=f"future lags ({region})"):
        trace = np.load(trace_l[i])
        dff   = np.load(dff_l[i])
        coord = np.load(coord_l[i])
        bind  = np.load(bind_l[i])

        name = adfn.save_name(dff_l[i])
        try:
            age = os.path.basename(dff_l[i]).split("dpf")[0][-1]
        except Exception:
            age = np.nan
        age_num = pd.to_numeric(age, errors="coerce")

        _, sub_dff, _, _ = tfn.select_region(trace, dff, bind, coord, region)
        X = sub_dff
        N, T = X.shape

        for lag_s in lag_secs:
            k = int(np.round(lag_s / dt))   # frames
            if k < 1:
                k = 1
            if k >= T:
                continue

            cos_d, mag_d = lag_cosine_and_magnitude(
                X, k=k,
                mag=mag,
                nan_on_zeros=nan_on_zeros,
                normalize_by_n=normalize_by_n
            )

            rows.append({
                "i": i,
                "name": name,
                "age": age,
                "age_num": age_num,
                "region": region,
                "lag_s": float(lag_s),
                "lag_frames": int(k),
                "cos_mean": float(np.nanmean(cos_d)),
                "mag_mean": float(np.nanmean(mag_d)),
                "n_neurons": int(N),
                "T": int(T),
                "dt": float(dt),
            })

    df_lags = pd.DataFrame(rows)
    df_lags = df_lags[np.isfinite(pd.to_numeric(df_lags["age_num"], errors="coerce"))].copy()
    df_lags["age_num"] = df_lags["age_num"].astype(float)
    return df_lags


# -----------------------------
# Usage
# -----------------------------
# dt = 1.0 / frame_rate_hz   # set this
df_future = build_future_lag_df(
    dff_l, trace_l, coord_l, bind_l,
    region="Telencephalon",
    dt=1/2.73,
    max_lag_s=10,
    step_s=1,
    mag="l2",
    nan_on_zeros=True,
    normalize_by_n=True
)

print(df_future.head())
print(df_future.shape)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from scipy.stats import spearmanr

def plot_age_vs_metric_coloured_by_lag_with_sig(
    df_future: pd.DataFrame,
    metric: str = "cos_mean",
    region: str = "Telencephalon",
    jitter: float = 0.08,
    seed: int = 0,
    print_table: bool = True,
):
    df = df_future.copy()
    df["age_num"] = pd.to_numeric(df["age_num"], errors="coerce")
    df["lag_s"]   = pd.to_numeric(df["lag_s"], errors="coerce")
    df[metric]    = pd.to_numeric(df[metric], errors="coerce")
    df = df[np.isfinite(df["age_num"]) & np.isfinite(df["lag_s"]) & np.isfinite(df[metric])].copy()

    if "region" in df.columns and region is not None:
        df = df[df["region"] == region].copy()

    # ensure one point per fish per lag (collapse any duplicates)
    df_fish = (df.groupby(["name", "age_num", "lag_s"], as_index=False)
                 .agg(y=(metric, "mean")))

    lags = np.sort(df_fish["lag_s"].unique())
    norm = Normalize(vmin=float(np.min(lags)), vmax=float(np.max(lags)))
    rng = np.random.default_rng(seed)

    # ---- Spearman per lag (fish-level) ----
    stats = []
    for lag in lags:
        sub = df_fish[df_fish["lag_s"] == lag]
        x = sub["age_num"].to_numpy(dtype=float)
        y = sub["y"].to_numpy(dtype=float)
        m = np.isfinite(x) & np.isfinite(y)
        n = int(m.sum())
        if n >= 3:
            rho, p = spearmanr(x[m], y[m])
        else:
            rho, p = np.nan, np.nan
        stats.append({"lag_s": float(lag), "n_fish": n, "rho": rho, "p": p})

    stats_df = pd.DataFrame(stats).sort_values("lag_s")

    if print_table:
        print(f"\nSpearman (fish-level) for metric='{metric}', region='{region}':")
        print(stats_df.to_string(index=False, float_format=lambda v: f"{v:.3g}"))

    # ---- Plot ----
    fig, ax = plt.subplots(figsize=(8, 5))

    xj = df_fish["age_num"].to_numpy() + rng.uniform(-jitter, jitter, size=len(df_fish))

    sc = ax.scatter(
        xj, df_fish["y"].to_numpy(),
        c=df_fish["lag_s"].to_numpy(),
        cmap="viridis", norm=norm,
        s=55, alpha=0.85,
        edgecolor="k", linewidths=0.2
    )

    # overlay mean trend per lag
    agg = (df_fish.groupby(["age_num", "lag_s"], as_index=False)
                 .agg(m=("y", "mean"))
                 .sort_values(["lag_s", "age_num"]))

    for lag in lags:
        sub = agg[agg["lag_s"] == lag].sort_values("age_num")
        if len(sub) >= 2:
            ax.plot(sub["age_num"].to_numpy(), sub["m"].to_numpy(),
                    alpha=0.55, linewidth=1.5)

    ax.set_xlabel("Age (dpf)")
    ax.set_ylabel(metric)
    ax.set_title(f"{region}: age vs {metric} (colour=lag s)\nSpearman printed per lag (fish-level)")

    cb = fig.colorbar(sc, ax=ax)
    cb.set_label("Lag (s)")

    fig.tight_layout()
    plt.show()

    return stats_df


# ---- run ----
stats_cos = plot_age_vs_metric_coloured_by_lag_with_sig(
    df_future, metric="cos_mean", region="Telencephalon", print_table=True)
```

```python
import os
import numpy as np
import pandas as pd

try:
    from tqdm.auto import tqdm
except Exception:
    def tqdm(x, **kwargs):
        return x


def build_neuron_timescale_df_with_fr(
    dff_l, trace_l, coord_l, bind_l, fr_l,
    region: str = "Telencephalon",
    dt: float = 1/2.73,
    max_lag_s: float = 5.0,
    threshold: float = 0.7,
    add_rate_residuals: bool = True,
    eps_fr: float = 1e-12,
):
    """
    Returns one row per neuron with:
      tau_s, acf1
      fr_hz (events/s)
      tau_per_event = tau_s * fr_hz
      (optional) residuals after regressing out fr within each fish:
        tau_resid_fr, acf1_resid_fr
    """
    dfs = []

    for i in tqdm(range(len(dff_l)), total=len(dff_l), desc=f"Timescales+FR ({region})"):
        trace = np.load(trace_l[i])
        dff   = np.load(dff_l[i])
        coord = np.load(coord_l[i])
        bind  = np.load(bind_l[i])

        name = adfn.save_name(dff_l[i])
        try:
            age = os.path.basename(dff_l[i]).split("dpf")[0][-1]
        except Exception:
            age = np.nan
        age_num = pd.to_numeric(age, errors="coerce")

        # ---- region select ----
        sub_trace, sub_dff, sub_bind, sub_coord = tfn.select_region(trace, dff, bind, coord, region)
        X = sub_dff  # (N,T)
        N, T = X.shape

        # ---- load firing rate(s) ----
        fr_obj = np.load(fr_l[i], allow_pickle=True).item()
        fr = fr_obj["fr"]

        # Case A: scalar fr for whole fish/region
        if np.isscalar(fr):
            fr_sub = np.full(N, float(fr), dtype=float)

        else:
            fr = np.asarray(fr, dtype=float)

            # Case B: per-neuron fr already matches telencephalon N
            if fr.ndim == 1 and fr.shape[0] == N:
                fr_sub = fr

            # Case C: fr is full-brain length; apply same region selection
            elif fr.ndim == 1 and fr.shape[0] == dff.shape[0]:
                fr_mat = fr[:, None]                       # (Nfull, 1)
                dummy  = np.zeros((dff.shape[0], 1), float) # dummy (Nfull, 1)
                sub_fr_mat, _, _, _ = tfn.select_region(fr_mat, dummy, bind, coord, region)
                fr_sub = np.asarray(sub_fr_mat[:, 0], dtype=float)
                if fr_sub.shape[0] != N:
                    raise ValueError(f"FR alignment failed: fr_sub {fr_sub.shape} vs sub_dff {X.shape}")

            else:
                raise ValueError(
                    f"Unsupported fr shape for i={i}: fr.shape={fr.shape}, "
                    f"sub_dff N={N}, full N={dff.shape[0]}. "
                    f"Expect scalar, (N_tel,), or (N_full,)."
                )

        # ---- compute tau + acf1 ----
        tau, acf1 = tau_half_life_from_acf_fast(
            X, dt=dt, max_lag_s=max_lag_s, threshold=threshold
        )
        tau = tau.astype(float)
        acf1 = acf1.astype(float)

        # ---- “normalize by rate” (dimensionless timescale in event units) ----
        tau_per_event = tau * fr_sub  # tau / (1/fr) = tau * fr

        # ---- build per-neuron df for this fish (vectorized, fast) ----
        df_i = pd.DataFrame({
            "i": i,
            "name": name,
            "age": age,
            "age_num": float(age_num) if np.isfinite(age_num) else np.nan,
            "region": region,
            "neuron_idx": np.arange(N, dtype=int),

            "tau_s": tau,
            "acf1": acf1,

            "fr_hz": fr_sub,
            "tau_per_event": tau_per_event,

            "n_neurons": int(N),
            "T": int(T),
            "dt": float(dt),
            "max_lag_s": float(max_lag_s),
            "threshold": float(threshold),
        })

        # ---- optional: regress out firing rate within fish (controls for rate without choosing a “normalization”) ----
        if add_rate_residuals:
            # Only use neurons with finite values
            m_tau = np.isfinite(df_i["tau_s"].to_numpy()) & np.isfinite(df_i["fr_hz"].to_numpy())
            m_acf = np.isfinite(df_i["acf1"].to_numpy())  & np.isfinite(df_i["fr_hz"].to_numpy())

            df_i["tau_resid_fr"] = np.nan
            df_i["acf1_resid_fr"] = np.nan

            # tau residuals
            if m_tau.sum() >= 10 and np.nanstd(df_i.loc[m_tau, "fr_hz"]) > eps_fr:
                x = df_i.loc[m_tau, "fr_hz"].to_numpy()
                y = df_i.loc[m_tau, "tau_s"].to_numpy()
                A = np.column_stack([np.ones_like(x), x])
                beta, *_ = np.linalg.lstsq(A, y, rcond=None)
                yhat = A @ beta
                df_i.loc[m_tau, "tau_resid_fr"] = y - yhat

            # acf1 residuals
            if m_acf.sum() >= 10 and np.nanstd(df_i.loc[m_acf, "fr_hz"]) > eps_fr:
                x = df_i.loc[m_acf, "fr_hz"].to_numpy()
                y = df_i.loc[m_acf, "acf1"].to_numpy()
                A = np.column_stack([np.ones_like(x), x])
                beta, *_ = np.linalg.lstsq(A, y, rcond=None)
                yhat = A @ beta
                df_i.loc[m_acf, "acf1_resid_fr"] = y - yhat

        dfs.append(df_i)

    df_neuron_tau_fr = pd.concat(dfs, ignore_index=True)
    df_neuron_tau_fr = df_neuron_tau_fr[np.isfinite(df_neuron_tau_fr["age_num"])].copy()
    df_neuron_tau_fr["age_num"] = df_neuron_tau_fr["age_num"].astype(float)
    return df_neuron_tau_fr


# -----------------------------
# Usage
# -----------------------------
df_neuron_tau_fr = build_neuron_timescale_df_with_fr(
    dff_l, trace_l, coord_l, bind_l, fr_l,
    region="Telencephalon",
    dt=1/2.73,
    max_lag_s=5,
    threshold=0.7,
    add_rate_residuals=True,   # strongly recommended
)

df_neuron_tau_fr.head()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

def fish_level_acf1_rate_normalised(df_neuron_tau_fr: pd.DataFrame, region="Telencephalon"):
    df = df_neuron_tau_fr.copy()
    df = df[df["region"] == region].copy()

    df["age_num"] = pd.to_numeric(df["age_num"], errors="coerce")
    df["acf1"]    = pd.to_numeric(df["acf1"], errors="coerce")
    df["fr_hz"]   = pd.to_numeric(df["fr_hz"], errors="coerce")
    df = df[np.isfinite(df["age_num"]) & np.isfinite(df["acf1"]) & np.isfinite(df["fr_hz"])].copy()

    # ---- Option A: ratio (less recommended, but you asked "normalised") ----
    df["acf1_div_fr"] = df["acf1"] / (df["fr_hz"] + 1e-12)

    # ---- Option B: residualise within fish: acf1 ~ fr_hz ----
    # (If you already computed acf1_resid_fr in the builder, we just use it.)
    if "acf1_resid_fr" not in df.columns or df["acf1_resid_fr"].isna().all():
        df["acf1_resid_fr"] = np.nan
        for (name, age), g in df.groupby(["name", "age_num"]):
            x = g["fr_hz"].to_numpy()
            y = g["acf1"].to_numpy()
            m = np.isfinite(x) & np.isfinite(y)
            if m.sum() < 10 or np.nanstd(x[m]) < 1e-12:
                continue
            A = np.column_stack([np.ones(m.sum()), x[m]])
            beta, *_ = np.linalg.lstsq(A, y[m], rcond=None)
            yhat = A @ beta
            resid = y[m] - yhat
            df.loc[g.index[m], "acf1_resid_fr"] = resid

    # ---- fish-level summaries ----
    fish = (df.groupby(["name", "age_num"], as_index=False)
              .agg(acf1_mean=("acf1", "mean"),
                   fr_mean=("fr_hz", "mean"),
                   acf1_div_fr_mean=("acf1_div_fr", "mean"),
                   acf1_resid_fr_mean=("acf1_resid_fr", "mean"),
                   n_neurons=("acf1", "size"))
              .sort_values(["age_num", "name"]))

    return fish


def _plot_fish_metric_vs_age_with_spearman(fish: pd.DataFrame, ycol: str, title: str):
    x = fish["age_num"].to_numpy(dtype=float)
    y = fish[ycol].to_numpy(dtype=float)
    m = np.isfinite(x) & np.isfinite(y)

    rho, p = (np.nan, np.nan)
    if m.sum() >= 3:
        rho, p = spearmanr(x[m], y[m])

    plt.figure(figsize=(7.5, 4.8))
    plt.scatter(x[m], y[m], s=70, alpha=0.85, edgecolor="k", linewidths=0.3)

    # per-age mean ± SEM (over fish)
    age_summary = (fish[m].groupby("age_num", as_index=False)
                   .agg(mean=(ycol, "mean"),
                        sem=(ycol, lambda v: v.std(ddof=1)/np.sqrt(len(v)) if len(v)>1 else np.nan),
                        n=("name", "size"))
                   .sort_values("age_num"))

    plt.errorbar(age_summary["age_num"], age_summary["mean"], yerr=age_summary["sem"],
                 fmt='-o', capsize=3)

    plt.xlabel("Age (dpf)")
    plt.ylabel(ycol)
    plt.title(f"{title}\nSpearman (fish-level): rho={rho:.3g}, p={p:.3g}")
    plt.tight_layout()
    plt.show()

    return rho, p, age_summary


# -------------------------
# Run
# -------------------------
fish = fish_level_acf1_rate_normalised(df_neuron_tau_fr, region="Telencephalon")

# Plot raw acf1 mean vs age
_plot_fish_metric_vs_age_with_spearman(
    fish, "acf1_mean",
    title="Telencephalon: mean lag-1 autocorr (raw) vs age"
)

# Plot 'normalised' by event rate (ratio) vs age
_plot_fish_metric_vs_age_with_spearman(
    fish, "acf1_div_fr_mean",
    title="Telencephalon: mean (acf1 / fr) vs age (simple normalisation)"
)

# Plot 'rate-controlled' residual vs age (recommended)
_plot_fish_metric_vs_age_with_spearman(
    fish, "acf1_resid_fr_mean",
    title="Telencephalon: mean acf1 residual after regressing out event rate (within fish) vs age"
)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

def fish_level_acf1_with_fr(df_neuron_tau_fr: pd.DataFrame, region="Telencephalon"):
    df = df_neuron_tau_fr.copy()
    df = df[df["region"] == region].copy()

    df["age_num"] = pd.to_numeric(df["age_num"], errors="coerce")
    df["acf1"]    = pd.to_numeric(df["acf1"], errors="coerce")
    df["fr_hz"]   = pd.to_numeric(df["fr_hz"], errors="coerce")
    df = df[np.isfinite(df["age_num"]) & np.isfinite(df["acf1"]) & np.isfinite(df["fr_hz"])].copy()

    fish = (df.groupby(["name", "age_num"], as_index=False)
              .agg(acf1_mean=("acf1", "mean"),
                   acf1_median=("acf1", "median"),
                   fr_mean=("fr_hz", "mean"),
                   n_neurons=("acf1", "size"))
              .sort_values(["age_num", "name"]))
    return fish

def add_fish_level_fr_residuals(fish: pd.DataFrame, ycol="acf1_mean", xcol="fr_mean"):
    fish = fish.copy()
    x = fish[xcol].to_numpy(dtype=float)
    y = fish[ycol].to_numpy(dtype=float)
    m = np.isfinite(x) & np.isfinite(y)

    fish["acf1_resid_fr_fish"] = np.nan
    if m.sum() >= 3 and np.nanstd(x[m]) > 1e-12:
        A = np.column_stack([np.ones(m.sum()), x[m]])
        beta, *_ = np.linalg.lstsq(A, y[m], rcond=None)
        yhat = A @ beta
        fish.loc[m, "acf1_resid_fr_fish"] = y[m] - yhat
    return fish

def plot_fish_metric_vs_age_with_spearman(fish: pd.DataFrame, ycol: str, title: str):
    x = fish["age_num"].to_numpy(dtype=float)
    y = fish[ycol].to_numpy(dtype=float)
    m = np.isfinite(x) & np.isfinite(y)

    rho, p = (np.nan, np.nan)
    if m.sum() >= 3:
        rho, p = spearmanr(x[m], y[m])

    plt.figure(figsize=(7.5, 4.8))
    plt.scatter(x[m], y[m], s=70, alpha=0.85, edgecolor="k", linewidths=0.3)

    age_summary = (fish[m].groupby("age_num", as_index=False)
                   .agg(mean=(ycol, "mean"),
                        sem=(ycol, lambda v: v.std(ddof=1)/np.sqrt(len(v)) if len(v)>1 else np.nan),
                        n=("name", "size"))
                   .sort_values("age_num"))

    plt.errorbar(age_summary["age_num"], age_summary["mean"], yerr=age_summary["sem"],
                 fmt='-o', capsize=3)

    plt.xlabel("Age (dpf)")
    plt.ylabel(ycol)
    plt.title(f"{title}\nSpearman (fish-level): rho={rho:.3g}, p={p:.3g}")
    plt.tight_layout()
    plt.show()
    return rho, p, age_summary

# -------------------------
# Run
# -------------------------
fish = fish_level_acf1_with_fr(df_neuron_tau_fr, region="Telencephalon")
fish = add_fish_level_fr_residuals(fish, ycol="acf1_mean", xcol="fr_mean")

# Raw ACF1 vs age
plot_fish_metric_vs_age_with_spearman(
    fish, "acf1_mean",
    title="Telencephalon: fish mean ACF1 vs age"
)

# FR-controlled residual vs age (THIS replaces your NaN plot)
plot_fish_metric_vs_age_with_spearman(
    fish, "acf1_resid_fr_fish",
    title="Telencephalon: fish ACF1 residual after regressing out fish firing rate vs age"
)

# Optional: print how strong ACF1~FR is (across fish)
m = np.isfinite(fish["acf1_mean"]) & np.isfinite(fish["fr_mean"])
if m.sum() >= 3:
    rho_fr, p_fr = spearmanr(fish.loc[m, "fr_mean"], fish.loc[m, "acf1_mean"])
    print(f"Spearman across fish: fr_mean vs acf1_mean rho={rho_fr:.3g}, p={p_fr:.3g}")

```

```python
# --- drop-in replacement: same function, but no right-side text labels ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

try:
    from tqdm.auto import tqdm
except Exception:
    def tqdm(x, **kwargs):
        return x


def _choose_extremes_by_percentile(tau: np.ndarray, p: float):
    tau = np.asarray(tau, dtype=float)
    ok = np.isfinite(tau)
    if ok.sum() < 5:
        return np.array([], dtype=int), np.array([], dtype=int), np.nan, np.nan

    lo = np.nanpercentile(tau[ok], p)
    hi = np.nanpercentile(tau[ok], 100.0 - p)

    fast_idx = np.where(ok & (tau <= lo))[0]
    slow_idx = np.where(ok & (tau >= hi))[0]
    return fast_idx, slow_idx, lo, hi


def plot_extreme_neurons_by_age(
    df_neuron_tau: pd.DataFrame,
    dff_l, trace_l, coord_l, bind_l,
    region: str = "Telencephalon",
    ages_to_show=(3, 4, 6, 8),
    percentile: float = 1.0,
    n_show_each: int = 3,
    use: str = "tau_s",            # "tau_s" or "acf1"
    signal: str = "dff",           # "dff" or "trace"
    tmax_s: float = 60.0,
    offset: float = None,
    seed: int = 0,
    linewidth: float = 1.0,
):
    rng = np.random.default_rng(seed)

    df = df_neuron_tau.copy()
    df["age_num"] = pd.to_numeric(df["age_num"], errors="coerce")
    df[use] = pd.to_numeric(df[use], errors="coerce")
    df = df[np.isfinite(df["age_num"]) & np.isfinite(df[use])].copy()

    if "region" in df.columns and region is not None:
        df = df[df["region"] == region].copy()

    ages_to_show = [float(a) for a in ages_to_show]
    df = df[df["age_num"].isin(ages_to_show)].copy()
    if df.empty:
        raise ValueError("No rows after filtering. Check ages_to_show/region/column names.")

    dt_default = float(pd.to_numeric(df.get("dt", pd.Series([np.nan])), errors="coerce").dropna().iloc[0])

    picks = []
    for age in ages_to_show:
        sub = df[df["age_num"] == age].copy()
        vals = sub[use].to_numpy(dtype=float)

        fast_local_idx, slow_local_idx, lo, hi = _choose_extremes_by_percentile(vals, percentile)
        if len(fast_local_idx) == 0 or len(slow_local_idx) == 0:
            print(f"[warn] age={age}: not enough finite values for extremes.")
            continue

        fast_take = rng.choice(fast_local_idx, size=min(n_show_each, len(fast_local_idx)), replace=False)
        slow_take = rng.choice(slow_local_idx, size=min(n_show_each, len(slow_local_idx)), replace=False)

        for idx in fast_take:
            r = sub.iloc[int(idx)]
            picks.append({"age_num": age, "kind": "fast", "i": int(r["i"]), "neuron_idx": int(r["neuron_idx"]), "val": float(r[use])})
        for idx in slow_take:
            r = sub.iloc[int(idx)]
            picks.append({"age_num": age, "kind": "slow", "i": int(r["i"]), "neuron_idx": int(r["neuron_idx"]), "val": float(r[use])})

    if len(picks) == 0:
        raise ValueError("No neurons picked. Try increasing percentile or loosening filters.")

    picks_df = pd.DataFrame(picks).sort_values(["age_num", "kind", "val"], ascending=[True, True, True])

    ages_present = sorted(picks_df["age_num"].unique())
    nrows = len(ages_present)
    fig, axes = plt.subplots(nrows=nrows, ncols=2, figsize=(12, 3.2*nrows), sharex=False, sharey=False)
    if nrows == 1:
        axes = np.array([axes])

    for r, age in enumerate(ages_present):
        for c, kind in enumerate(["fast", "slow"]):
            ax = axes[r, c]
            subp = picks_df[(picks_df["age_num"] == age) & (picks_df["kind"] == kind)].copy()
            if subp.empty:
                ax.axis("off")
                continue

            traces, dts = [], []
            for _, row in subp.iterrows():
                i = int(row["i"])
                neuron_idx = int(row["neuron_idx"])

                trace = np.load(trace_l[i])
                dff   = np.load(dff_l[i])
                coord = np.load(coord_l[i])
                bind  = np.load(bind_l[i])

                sub_trace, sub_dff, _, _ = tfn.select_region(trace, dff, bind, coord, region)
                Xsig = sub_dff if signal.lower() == "dff" else sub_trace
                x = Xsig[neuron_idx].astype(float)

                dt_i = dt_default
                if "dt" in df_neuron_tau.columns:
                    dt_rows = df_neuron_tau.loc[df_neuron_tau["i"] == i, "dt"]
                    if len(dt_rows):
                        v = pd.to_numeric(dt_rows.iloc[0], errors="coerce")
                        if np.isfinite(v):
                            dt_i = float(v)

                n_keep = int(min(len(x), np.floor(tmax_s / dt_i)))
                n_keep = max(10, n_keep)
                traces.append(x[:n_keep])
                dts.append(dt_i)

            if offset is None:
                amp = np.nanmedian([np.nanpercentile(t, 95) - np.nanpercentile(t, 5) for t in traces])
                off = 1.2 * (amp if np.isfinite(amp) and amp > 0 else 1.0)
            else:
                off = float(offset)

            for j, (x, dt_i) in enumerate(zip(traces, dts)):
                t = np.arange(len(x)) * dt_i
                ax.plot(t, x + j*off, linewidth=linewidth)

            ax.set_title(f"age={int(age)} dpf | {kind} ({signal}) | extremes={percentile}% | n={len(traces)}")
            ax.set_xlabel("Time (s)")
            ax.set_ylabel("Signal (offset)")
            ax.grid(False)

    fig.suptitle(f"{region}: example single-neuron traces for {use} extremes", y=1.02, fontsize=13)
    fig.tight_layout()
    plt.show()

    return picks_df


# Example:
picks_df = plot_extreme_neurons_by_age(df_neuron_tau, dff_l, trace_l, coord_l, bind_l,
                                       ages_to_show=(3,8), percentile=1.0, n_show_each=10,
                                       use="tau_s", signal="dff", tmax_s=60)
```

```python
# --- only change is the title line: show age only (no name) ---

def plot_fish_mean_activity_grid(
    df_mean_ts: pd.DataFrame,
    ages=(3,4,5,6,7,8),
    region="Telencephalon",
    signal="dff",
    n_cols: int = 3,
    max_points: int = 2000,
):
    df = df_mean_ts.copy()
    df = df[(df["region"] == region) & (df["signal"] == signal)].copy()
    df = df[df["age_num"].isin([float(a) for a in ages])].copy()

    fish = (df[["name", "age_num"]].drop_duplicates()
              .sort_values(["age_num", "name"])
              .reset_index(drop=True))

    n = len(fish)
    n_rows = int(np.ceil(n / n_cols))

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(4.2*n_cols, 2.6*n_rows), sharex=False, sharey=False)
    axes = np.array(axes).reshape(n_rows, n_cols)

    for ax in axes.ravel():
        ax.axis("off")

    for idx, row in fish.iterrows():
        r = idx // n_cols
        c = idx % n_cols
        ax = axes[r, c]
        ax.axis("on")

        name = row["name"]
        age  = row["age_num"]
        sub = df[df["name"] == name].sort_values("t")

        t = sub["t"].to_numpy()
        y = sub["mean_activity"].to_numpy()

        if len(t) > max_points:
            ii = np.linspace(0, len(t)-1, max_points).astype(int)
            t = t[ii]
            y = y[ii]

        ax.plot(t, y, linewidth=1.0)
        ax.set_title(f"age={int(age)} dpf", fontsize=9)  # <-- no name
        ax.set_xlabel("t (s)" if sub["t"].max() > 100 else "t")
        ax.set_ylabel("mean activity")

    fig.suptitle(f"{region}: mean activity across neurons ({signal}) — one panel per fish", y=1.02)
    fig.tight_layout()
    plt.show()

    return fish


# run:
fish_index = plot_fish_mean_activity_grid(df_mean_ts, ages=(3,4,5,6,7,8), region="Telencephalon", signal="dff")
```
